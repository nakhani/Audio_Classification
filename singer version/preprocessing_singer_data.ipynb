```python
import os
import pydub
import matplotlib.pyplot as plt
import pydub.silence
import pydub.utils
from pydub.utils import make_chunks
from spleeter.separator import Separator
```

```python
separator = Separator('spleeter:2stems')

files = os.listdir("/content/drive/MyDrive/dataset/vocal_data")
merged_file = []
all_data = []

for file in files:
    if "_" in file:
        first_part_name = file.split("_")[0] 


     
        matched_file = None
        for merged in merged_file:
            if merged.startswith(first_part_name) and "_" in merged:
                matched_file = merged
                break

        if matched_file:
            
            voice_1 = pydub.AudioSegment.from_file(os.path.join("/content/drive/MyDrive/dataset/vocal_data", matched_file))
            voice_2 = pydub.AudioSegment.from_file(os.path.join("/content/drive/MyDrive/dataset/vocal_data", file))
            result = voice_1 + voice_2

           
            merged_file_path = os.path.join("/content/drive/MyDrive/dataset/vocal_data" , (first_part_name + f".wav"))
            result.export(merged_file_path)
            all_data.append(merged_file_path)
            print(f"Merged {matched_file} and {file} into {first_part_name}")

            
            merged_file.remove(matched_file)
        else:
            
            merged_file.append(file)
    else:
    
        all_data.append(os.path.join("/content/drive/MyDrive/dataset/vocal_data", file))
```

```python
separator = Separator('spleeter:2stems')

vocal_data_dir = "/content/drive/MyDrive/dataset/vocal_data"


all_data = [os.path.join(vocal_data_dir, file) for file in os.listdir(vocal_data_dir) if file.endswith(".wav")]

for data in  all_data:

    separator.separate_to_file(data, 'spleeter_output')
    vocal_file_path = os.path.join('spleeter_output', os.path.splitext(os.path.basename(data))[0], 'vocals.wav')

    audio = pydub.AudioSegment.from_file(vocal_file_path)
    audio = audio.set_sample_width(2)
    audio = audio.set_frame_rate(48000)
    audio = audio.set_channels(1)

    file_name = os.path.basename(data).split(".")[0]

    os.makedirs(os.path.join("/content/drive/MyDrive/dataset/final_data_singer",file_name), exist_ok=True)


    # audio_array = audio.get_array_of_samples()
    # plt.figure(figsize=(20,2))
    # plt.plot(audio_array)
    # plt.show()

    chunks = pydub.silence.split_on_silence(audio, min_silence_len=3000  , silence_thresh=-40)

    if chunks:  
        result = sum(chunks)
        parts = make_chunks(result, 1000)
        for i, part in enumerate(parts):
          if len(part) >= 1000:
           part.export(os.path.join("/content/drive/MyDrive/dataset/final_data_singer",file_name, f"voice_{i}.wav"), format="wav")
           print(f"Processed and exported voice_{i}.wav")
    else:
        parts = make_chunks(audio, 1000)
        for i, part in enumerate(parts):
          if len(part) >= 1000:
           part.export(os.path.join("/content/drive/MyDrive/dataset/final_data_singer",file_name, f"voice_{i}.wav"), format="wav")
           print(f"Processed and exported voice_{i}.wav")
```
